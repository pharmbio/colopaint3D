```python
df4 = pd.read_csv('Layout_Files/DrugDict_exp180.csv', sep=';', index_col=0) 
```

```python
df5 = df4[~df4['Dro'].str.contains('AE')]
```

```python
df6 = df5[~df5['Dro'].str.contains('AF')]
```

```python
df6.reset_index(drop=True, inplace=True)
```

```python
from string import ascii_uppercase as letters
for letter in letters:
    for indI in range(1,10):
        print('preop', letter+str(indI))
        print('postop', letter+'0'+str(indI))
        df6['DestWell'].replace(letter+str(indI), letter+'0'+str(indI), inplace=True)
```

```python
for indI in range(1, 10):
    df6['DestWell'].replace('AA'+str(indI), 'AA0'+str(indI), inplace=True)
    df6['DestWell'].replace('AB'+str(indI), 'AB0'+str(indI), inplace=True)
    df6['DestWell'].replace('AC'+str(indI), 'AC0'+str(indI), inplace=True)
    df6['DestWell'].replace('AD'+str(indI), 'AD0'+str(indI), inplace=True)
```

```python
df6.to_csv('Layout_Files/DrugDict_exp180.csv', sep=';')
```
