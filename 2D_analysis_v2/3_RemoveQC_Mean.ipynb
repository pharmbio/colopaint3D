## Read Feature and Layout data 
Metadata matters! 

```python
# Feature data 
BaseDir = "./ImageMeanFeatures/"
df = pd.read_csv("{}/ImageMeansAllPlates.csv".format(BaseDir), sep=";", index_col=0)
df.head(1)
```

```python
# Some necessary adaptations on the dataframe
df.drop(['Metadata_Site_nuclei','Metadata_Site_cytoplasm','Metadata_Site_cells','ImageNumber_cells','ImageNumber_cytoplasm', 'ImageNumber_nuclei'], axis=1, inplace=True) # "ImageNumber_nuclei": is 1 across the entire plate, it originates in the pipelineGUI where we run batches of 1 image only -> every image will be no. 1, "Metadata_Site_nuclei": have a new column now. 
df["plate_well"] = df.index.str[0:-2] # Well identifier accross an entire batch
df.head(1)
```

```python
List = df.columns.values
```

## Remove QC flagged images

```python
QC = pd.read_csv("./QC/colo44_Flags.csv", sep=";", index_col=0)
QC["ImageID"] = (
    QC["Metadata_Barcode"]
    + "_"
    + QC["Metadata_Well"]
    + "_"
    + QC["Metadata_Site"].astype(str)
)
QC.head(3)
```

```python
dfQC = df.merge(QC[['ImageID', 'Total', 'Count_nuclei']], left_on = 'ImageID', right_on = 'ImageID', how = 'left') 
dfQC = dfQC[dfQC['Total'] == 0]
dfQC.drop(["Total"], axis=1, inplace=True)
dfQC.reset_index(drop=True, inplace=True)
dfQC.head()
```

## Save the combined data

```python
dfQC.to_csv("{}/colo44_QC_filtered_Mean.csv".format(OutputDir), sep=";")
```
