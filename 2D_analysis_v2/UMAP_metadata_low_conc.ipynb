```python
import os
import umap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.text import Text
import seaborn as sns; sns.set_style("white")
import umap.plot
%matplotlib inline
os.getcwd()
```

```python
def pDose(x):
    '''This is just a helper function, to compute easily log transformed concentrations used in drug discovery'''
    return(-np.log10(1e-6*x))
```

```python
BaseDir = "./results/"
```

```python
# Import feature data after QC
selected_df = pd.read_csv("{}colopaint44_selected_data_notMAD.csv".format(BaseDir), sep=";", index_col=0)
selected_df.head(1)
```

```python
df_toplot = selected_df
df_toplot['Metadata_cmpd_conc'].loc[(df_toplot['Metadata_cmpd'] == '[dmso]')] = 0.0001
df_toplot['Metadata_inv_log_conc'] = 10 - pDose(df_toplot['Metadata_cmpd_conc'])
ListOfMetadata = list(df_toplot.columns[df_toplot.columns.str.contains("Metadata_")])
```

```python
subsetDf = df_toplot[df_toplot['Metadata_cmpd_conc'] != 0.00833]
subsetDf = subsetDf[subsetDf['Metadata_cmpd_conc'] != 0.03330] # Remove lowest concentrations
subsetDf = subsetDf[subsetDf['Metadata_cmpd_conc'] != 0.1]
#subsetDf = subsetDf[subsetDf['Metadata_cmpd_conc'] != 0.333]
training_data_subset = subsetDf.drop(ListOfMetadata, axis=1).to_numpy()
```

```python
## UMAP - MOA

reducer = umap.UMAP(n_components=2, n_neighbors=23, min_dist= 0.3, random_state=42)
embedding = reducer.fit_transform(training_data_subset)

fig = plt.figure()
ax = sns.scatterplot(
    x=embedding[:, 0],
    y=embedding[:, 1],
    hue=subsetDf.Metadata_moa,
    alpha=(0.7),
    marker="o",
    #size=df_toplot["Metadata_cmpd_conc"],
    #palette=pal,
    #legend=False
    )
plt.legend(bbox_to_anchor=(2.8, 1),loc='upper right')
plt.title('Moa')
```

```python
## UMAP - Class
fig = plt.figure()
ax = sns.scatterplot(
    x=embedding[:, 0],
    y=embedding[:, 1],
    hue=subsetDf.Metadata_class,
    alpha=(0.7),
    marker="o",
    #size=df_toplot["Metadata_cmpd_conc"],
    #palette=pal,
    #legend=False
    )

plt.title('Class')
```

```python
## UMAP - Compound
fig = plt.figure()
ax = sns.scatterplot(
    x=embedding[:, 0],
    y=embedding[:, 1],
    hue=subsetDf.Metadata_cmpd,
    alpha=(0.7),
    marker="o",
    #size=df_toplot["Metadata_cmpd_conc"],
    #palette=pal,
    legend=False
    )
plt.title('Compound')
```
